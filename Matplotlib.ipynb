```python
# 1st Method
fig=plt.figure()# creates a figure
ax=fig.add_subplot()# adds axes
plt.plot()
```

```python
#2nd Method
fig=plt.figure()
ax=fig.add_axes([1,1,1,1])
ax.plot(x,y)# adds data
plt.plot()
```

```python
#3rd Method (Most Recommended)
fig,ax=plt.subplots()# resets the figure and inserts new inserted data
ax.plot(x,[50,60,44,54,55])
plt.plot()
```

# Matplotlib example workflow

```python
# 1) Import Matplot lib and get it ready for plotting
%matplotlib inline
import matplotlib.pyplot as plt

# 2) Prepare Data
x=[1,2,3,4,5]
y=[6,7,8,9,10]

# 3) setup plot
fig,ax= plt.subplots(figsize=(10,5))# figsize= (width, height)

#4) plot data
ax.plot(x,y)

#5) customize plot
ax.set(title='sample plot',xlabel='x-axis',ylabel='y-axis') #setting up title and labels for x and y axes

#6) save and share
fig.savefig("../practice/numpy-images/savepic.png") # specify any path
```

```python
# Histogram
x=np.random.random(100)
fig,ax=plt.subplots()
ax.hist(x)
```

```python
x=np.random.randn(1000)# histogram for Normal distribution( here we used randn insted of random)
fig,ax=plt.subplots()
ax.hist(x)
```

# Heart Disease Example

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('heart-disease.csv')
df.head()
```

```python
df.plot.hist(figsize=(10,20),subplots=True); # For histograms of each column value as subplots
```

```python
over_50=df[df['age']>50] # Age of people over 50
over_50
```

```python
over_50.plot(kind='scatter',x='age',y='chol',c='red'); 
# kind--> defines which type of plot, whether it is scatter or hist or bar...
```

```python
# Another way of representing the above graph
over_50.plot(kind='scatter',x='age',y='chol',c='target',figsize=(10,6))
```

```python
over_50.plot(kind='scatter',x='age',y='chol',c='target',figsize=(10,6));
# semicolon will remove the text above the graph
```

```python
# Now we will use Object oriented method
fig,ax=plt.subplots(figsize=(10,6))
#plot the data
scatter=ax.scatter(x=over_50['age'],y=over_50['chol'],c='blue')
```

```python
fig,ax=plt.subplots(figsize=(10,6))

#plot the data
scatter=ax.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'])

#Customizing the plot
ax.set(title='Heart_Disease_Graph', xlabel='age',ylabel='chol')
```

```python
fig,ax=plt.subplots(figsize=(10,6))

#plot the data
scatter=ax.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'])

#Customizing the plot
ax.set(title='Heart_Disease_Graph', xlabel='age',ylabel='chol')

# Now Adding Legend to the Graph
ax.legend(*scatter.legend_elements(),title='target');

# Adding a Horizontal Line in the plot
ax.axhline(over_50['chol'].mean(),linestyle='-.')
```

```python
fig,ax=plt.subplots(figsize=(10,6))

#plot the data
scatter=ax.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'])

#Customizing the plot
ax.set(title='Heart_Disease_Graph', xlabel='age',ylabel='chol')

# Now Adding Legend to the Graph
ax.legend(*scatter.legend_elements(),title='target');

# Adding a Vertical Line in the plot
ax.axvline(over_50['age'].mean(),linestyle='-.')
```

```python
over_50.head()
```

```python
# Viewing it as Subplots
fig, (ax0,ax1)=plt.subplots(2,1,figsize=(10,10))# syntax for subplot

#Customizing axis0
scatter=ax0.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'])

ax0.set(title='Heart_Disease vs Cholestrol', xlabel='age',ylabel='chol')

ax0.legend(*scatter.legend_elements(),title='target')

ax0.axhline(over_50['chol'].mean(),linestyle='-.');

#Customizing axis1
scatter=ax1.scatter(x=over_50['age'],y=over_50['thalach'],c=over_50['target'])

ax1.set(title='Heart_Disease vs Max Heart rate', xlabel='age',ylabel='thalach')

ax1.legend(*scatter.legend_elements(),title='target')

ax1.axhline(over_50['thalach'].mean(),linestyle='-.');

#Adding a super title for both the plots combined
fig.suptitle('Heart Disease Check',fontsize=20);
```
